```python
import json

def create_input_prompt(system_message, user_prompt):
    return {
        "messages": [
            {"role": "system","content": system_message},
            {"role": "user", "content": user_prompt},
        ]
    }

# Open and read the JSON file
with open("named_entity_class_dictionary.json", "r") as json_file:
    named_entity_classes_dict = json.load(json_file)

# Print the loaded dictionary
print(named_entity_classes_dict)
named_entity_classes = [named_entity_class for named_entity_class in named_entity_classes_dict]
print(named_entity_classes)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def most_similar_sentence(test_sentence, dataset):
    # Split the dataset into individual sentences
    sentences_dataset = dataset

    # Create TF-IDF vectors for the test sentence and dataset sentences
    vectorizer = TfidfVectorizer().fit([test_sentence] + sentences_dataset)
    tfidf_test_sentence = vectorizer.transform([test_sentence])
    tfidf_dataset = vectorizer.transform(sentences_dataset)

    # Compute the cosine similarity between the test sentence and all dataset sentences
    similarity_matrix = cosine_similarity(tfidf_dataset, tfidf_test_sentence)

    # Find the index of the sentence with the highest similarity
    most_similar_index = similarity_matrix.argmax()

    # Extract the most similar sentence
    most_similar_sentence = sentences_dataset[most_similar_index]

    return most_similar_sentence, most_similar_index
```

```python
num = 0
sample_sentence=""
sample_output=""
test_dataset = dict()
for sentence in sentences:
    num +=1
    example_sentence, example_sentence_index = most_similar_sentence(sentence, example_sentences)
    example_output = example_outputs[example_sentence_index]
    system_message = f"""Given the following entity classes and sentences, label entity mentions with their respective classes in sentences according to the sentences' context. In the output, only include entity mentions and their respective class in the given output format. No needed further explanation.\nCONTEXT: entity classes: {named_entity_classes}.\nExample sentence: {example_sentence}.\nExample output: {example_output.strip()} 
    """
    
    user_prompt = f"""
    Test sentence: {sentence}
    Test output:
    """
    
    message = create_input_prompt(system_message, user_prompt)
    test_dataset[num] = message
```

```python
import json
# Save the dictionary to a JSON file
with open("test_prompts.json", "w") as json_file:
    json.dump(test_dataset, json_file, indent=4)  # 'indent' adds formatting for readability
```
